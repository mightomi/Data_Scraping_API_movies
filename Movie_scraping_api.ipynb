```python
import pandas as pd
import numpy as np
import json
from urllib.request import urlopen
import requests
```

```python
url="https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page="
```

```python
json_page=requests.get(url+"").json()
json_page
```

```python
popularity=[]
vote=[]
id_=[]
title=[]
release_date=[]
overview=[]
```

```python
for i in range (1,11):

    json_page=requests.get(url+"2").json()
    
    for item in json_page["results"]:
        popularity.append(item["popularity"])
        vote.append(item["vote_average"])
        id_.append(item["id"])
        title.append(item["title"])
        overview.append(item["overview"])
        release_date.append(item["release_date"])
```

```python
movie_list=pd.DataFrame()

movie_list["Movie Id"]=id_
movie_list["Title"]=title
movie_list["Release Date"]=release_date
movie_list["Popularity"]=popularity
movie_list["Rating"]=vote
movie_list["Overview"]=overview

```
